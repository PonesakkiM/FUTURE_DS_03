```python
import pandas as pd
df = pd.read_csv("College_Fest_Feedback_Dataset.csv")
df.head()
```

Remove Duplicates and Null values

```python
df.drop_duplicates(inplace=True)
```

```python
df=df.dropna(subset=["Feedback","Event_Type","Role","Role_Details","Suggestions"])
```

Event ratings Distibution Graph

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='Rating', data=df)
plt.title("Distribution of Event Ratings")
plt.show()
```

Department wise Satisfaction

```python
dept_avg = df.groupby('Department')['Overall_Satisfaction'].mean().reset_index()
sns.barplot(x='Department', y='Overall_Satisfaction', data=dept_avg)
plt.title("Average Satisfaction by Department")
plt.show()
```

Satisfaction for event_type

```python
sns.boxplot(x='Event_Type', y='Overall_Satisfaction', data=df,color='darkblue')
plt.title("Satisfaction by Event Type")
plt.show()
```

Satisfaction level for Roles

```python
sns.boxplot(x='Role', y='Overall_Satisfaction', data=df, color='cyan',linecolor = 'darkblue')
plt.title("Satisfaction by Role")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
def get_sentiment(text):
    if pd.isna(text):
        return "Neutral"
    polarity = TextBlob(str(text)).sentiment.polarity
    if polarity > 0:
        return "Positive"
    elif polarity == 0:
        return "Neutral"
    else:
        return "Negative"
df['Sentiment'] = df['Feedback'].apply(get_sentiment)
sns.countplot(x='Sentiment', data=df,color="orange",edgecolor="green")
plt.title("Feedback Sentiment Distribution")
plt.show()
```

```python
from wordcloud import WordCloud

text = " ".join(df['Suggestions'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title("Word Cloud",color='red')
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
avg_ratings = df[['Rating', 'Organization_Rating', 'Overall_Satisfaction']].mean()
labels = ['Event Rating', 'Organization Rating', 'Overall Satisfaction']
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]
values = avg_ratings.tolist()
values += values[:1]
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, values, linewidth=2, color='darkblue')
ax.fill(angles, values, alpha=0.25, color='red')
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
plt.title("Overall Ratings Radar Chart", size=14)
plt.show()
```
